# Project goal: Predict Cereal Yield

### Purpose: Exploratory Data Analysis

## Data

The attributes/features are downloaded and scrap from several webpages, e.g., Climate Change Knowledge Portal and  World Bank Data.

Original number of countries/regions = 206. Remove those without data, the remaining number of countries are 111. Among 95 countries that are excluded, 45 of them have not cereal yield data.

##### The features are:

A: Agricultural arable land (% of land area) 

Rmin: Minimum precipitation in mm

Tavg: Mean air temperature in degree C 

R: Rural population (% of total population)

F: Fertilizer consumption (kilograms per hectare of arable land) 

M: Agricultural machinery, tractors per 100 sq. km of arable land 

V: Agriculture, forestry, and fishing, value added (% of GDP) 

Gc: GDP per capita (current US$) 

E :Employment in agriculture (% of total employment) 

Dl : Pump price for diesel fuel (US$ per liter) 


##### The target is:
Ckg: Cereal yield (kg per hectare)

```python
df_train = df[(df['year']>='1991') & (df['year']<='2010')]
df_test = df[(df['year']>'2010') & (df['year']<='2015')]
```

## Data Cleaning

```python
# Drop the features ['P', 'Gl', 'Cpl', 'G', 'Tmin', 'Tmax', 'Ravg', 'Rmax']
df_train.drop(columns=['P', 'Gl', 'Cpl', 'G', 'Tmin', 'Tmax', 'Ravg', 'Rmax'], inplace=True)
df_test.drop(columns=['P', 'Gl', 'Cpl', 'G','Tmin', 'Tmax', 'Ravg', 'Rmax'], inplace=True)
```

```python
# Apply imputation methods: Regression and Substitution
# Replace the missing values depending which years have missing data
# col_names = ['F', 'V', 'A', 'Ckg', 'Gc']
country_codes = df_train['code'].drop_duplicates().values
for c in country_codes:
    idx = df_train[df_train['code']==c].index
    idxt = df_test[df_test['code']==c].index
    
    # For V
    #  1995 is the year when all of the contries have data in the training data. 
    #    Any year before 1995 has no data, it will be substituted with the value in year 1995
    #    Any year after 1995 has no data, it will be substituted with the mean value of
    #       all the available data. 
    #  For test data, the missing values are replaced by the mean value calculated from the
    #       training data set
    df_temp = df_train.loc[idx][df_train.loc[idx]['V'].isnull()]
    df_notnull = df_train.loc[idx][df_train.loc[idx]['V'].notnull()]
    if not df_temp.empty:
        if not df_temp[df_temp['year']>'1995']['V'].empty:
            idx1 = df_temp[df_temp['year']>'1995']['V'].index
            df_train.loc[idx1,'V'] = df_notnull['V'].mean()
            
        idx2 = df_test.loc[idxt][df_test.loc[idxt]['V'].isnull()].index
        if not idx2.empty:
            df_test.loc[idx2,'V'] = df_notnull['V'].mean()
        
        if not df_temp[df_temp['year']<'1996']['V'].empty:
            idx1 = df_temp[df_temp['year']<'1996']['V'].index
            df_train.loc[idx1,'V'] = df_notnull['V'].iloc[0]        
    
    # Gc - Similar to V but the selected year is 2001
    df_temp = df_train.loc[idx][df_train.loc[idx]['Gc'].isnull()]
    df_notnull = df_train.loc[idx][df_train.loc[idx]['Gc'].notnull()]
    if not df_temp.empty:
        if not df_temp[df_temp['year']>'2001']['Gc'].empty:
            idx1 = df_temp[df_temp['year']>'2001']['Gc'].index
            df_train.loc[idx1,'Gc'] = df_notnull['Gc'].mean()

        idx2 = df_test.loc[idxt][df_test.loc[idxt]['Gc'].isnull()].index
        if not idx2.empty:
            df_test.loc[idx2,'Gc'] = df_notnull['Gc'].mean()
                
        if not df_temp[df_temp['year']<'2002']['Gc'].empty:
            idx1 = df_temp[df_temp['year']<'2002']['Gc'].index
            df_train.loc[idx1,'Gc'] = df_notnull['Gc'].iloc[0]
    
    
    # F, A
    sub_values = [df_train[df_train.code==c]['F'].min(), \
                  df_train[df_train.code==c]['A'].min()]
    
    df_train.loc[idx,'F'] = df_train.loc[idx,'F'].fillna(sub_values[0])
    df_test.loc[idxt,'F'] = df_test.loc[idxt,'F'].fillna(sub_values[0])
    df_train.loc[idx,'A'] = df_train.loc[idx,'A'].fillna(sub_values[1])
    df_test.loc[idxt,'A'] = df_test.loc[idxt,'A'].fillna(sub_values[1])
    
    # The missing data for the rest of the features are imputed with regression method
    df_train.loc[idx,'Dl'] = df_train.loc[idx,'Dl'].interpolate(method='linear', limit_direction="both")
    df_train.loc[idx,'M'] = df_train.loc[idx,'M'].interpolate(method='linear', limit_direction="both")
    df_train.loc[idx,'R'] = df_train.loc[idx,'R'].interpolate(method='linear', limit_direction="both")
    df_train.loc[idx,'Ckg'] = df_train.loc[idx,'Ckg'].interpolate(method='linear', limit_direction="both")
    
    # For the missing data in the test data set, just replace them with the value in the
    #   training data set
    df_test.loc[idxt,'Dl'] = df_test.loc[idxt,'Dl'].fillna(df_train.loc[idx,'Dl'].values[-1])
    df_test.loc[idxt,'M'] = df_test.loc[idxt,'M'].fillna(df_train.loc[idx,'M'].values[-1])
    df_test.loc[idxt,'R'] = df_test.loc[idxt,'R'].fillna(df_train.loc[idx,'R'].values[-1])
    df_test.loc[idxt,'Ckg'] = df_test.loc[idxt,'Ckg'].fillna(df_train.loc[idx,'Ckg'].values[-1])
    
```
